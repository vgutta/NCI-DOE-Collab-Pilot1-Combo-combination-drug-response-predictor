```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cell_almanac = ['NCI60_ALMANAC.tsv', 'NCIPDM_ALMANAC.tsv', 'GDSC_ALMANAC.tsv', 'RTS_ALMANAC.tsv']
cell_gdsc = ['NCI60_GDSC.tsv', 'NCIPDM_GDSC.tsv', 'GDSC_GDSC.tsv', 'RTS_GDSC.tsv']
cell_nci_ioa_aoa = ['NCI60_NCI_IOA_AOA.tsv', 'NCIPDM_NCI_IOA_AOA.tsv', 'GDSC_NCI_IOA_AOA.tsv', 'RTS_NCI_IOA_AOA.tsv']
cell_rts = ['NCI60_RTS.tsv', 'NCIPDM_RTS.tsv', 'GDSC_RTS.tsv', 'RTS_RTS.tsv']
```

### Generating DataFrames, one for each Drugset

```python
df_almanac = pd.read_csv('./cell_line/NCI60_ALMANAC.tsv', sep=',', usecols=['Drug1', 'Drug2'])
for file in cell_almanac:
    tmp_pd = pd.read_csv('./cell_line/' + file, sep=',')
    df_almanac[file.rpartition('_')[0]] = tmp_pd['PredGrowthMean']
   
tissue_almanac = pd.read_csv('./tissue/ALMANAC_TCGA-THCA.tsv', sep=',', usecols=['PredGrowthMean'])
df_almanac['TCGA_THCA'] = tissue_almanac['PredGrowthMean']


df_gdsc = pd.read_csv('./cell_line/NCI60_GDSC.tsv', sep=',', usecols=['Drug1', 'Drug2'])
for file in cell_gdsc:
    tmp_pd = pd.read_csv('./cell_line/' + file, sep=',')
    df_gdsc[file.rpartition('_')[0]] = tmp_pd['PredGrowthMean']
   
tissue_gdsc = pd.read_csv('./tissue/GDSC_TCGA-THCA.tsv', sep=',', usecols=['PredGrowthMean'])
df_gdsc['TCGA_THCA'] = tissue_gdsc['PredGrowthMean']


df_nci_ioa_aoa = pd.read_csv('./cell_line/RTS_NCI_IOA_AOA.tsv', sep=',', usecols=['Drug1', 'Drug2'])
for file in cell_nci_ioa_aoa:
    tmp_pd = pd.read_csv('./cell_line/' + file, sep=',')
    df_nci_ioa_aoa[file.partition('_')[0]] = tmp_pd['PredGrowthMean']

tissue_nci_ioa_aoa = pd.read_csv('./tissue/NCI_IOA_AOA_TCGA-THCA.tsv', sep=',', usecols=['PredGrowthMean'])
df_nci_ioa_aoa['TCGA_THCA'] = tissue_nci_ioa_aoa['PredGrowthMean']


df_rts = pd.read_csv('./cell_line/RTS_RTS.tsv', sep=',', usecols=['Drug1', 'Drug2'])
for file in cell_rts:
    tmp_pd = pd.read_csv('./cell_line/' + file, sep=',')
    df_rts[file.rpartition('_')[0]] = tmp_pd['PredGrowthMean']
   
tissue_rts = pd.read_csv('./tissue/RTS_TCGA-THCA.tsv', sep=',', usecols=['PredGrowthMean'])
df_rts['TCGA_THCA'] = tissue_rts['PredGrowthMean']
```

### Plots
Combo scores for a combination of drugs and a sample are the mean of 10 predictions Combo model made

#### ALMANAC Dataframe and Plots

```python
df_almanac[['NCI60', 'NCIPDM', 'GDSC', 'RTS', 'TCGA_THCA']].plot(style=".",
         title="ALMANAC Drug sampleset",
         xlabel="Index positions of combinations of drugs from sampleset",
         ylabel="Combo Score",
         figsize=(15,10))
```
